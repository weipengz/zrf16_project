### Signal Generation

```python
## Generate signal
num_symbols = 16
bits1 = np.random.randint(0, 2, num_symbols)*2-1 # Our data to be transmitted, 1's and 0's
bits2 = np.random.randint(0, 2, num_symbols)*2-1 # Our data to be transmitted, 1's and 0's
print(bits1, bits2)
```

```python
sps = 8
wfm1 = helper.pulse_shaping(bits1, sps)
wfm2 = helper.pulse_shaping(bits2, sps)

mix = np.array([1,1])
wfm_weighted = np.dot(np.array([wfm1,wfm2]).T,mix)

plt.figure(figsize=(10,3))
n_cat = 10
plt.plot(np.concatenate([wfm_weighted[-n_cat:],wfm_weighted,wfm_weighted[:n_cat]]), '.-')
for i in range(num_symbols):
    plt.plot([i*sps+n_cat,i*sps+n_cat], [0, wfm_weighted[i*sps]])
plt.grid(True)
plt.show()
```

```python
data_length = 256
np1 = 10
np2 = 10
np3 = 20
wfm1 = np.sin(np1*2*np.pi*np.linspace(0,1,data_length,endpoint=False))
wfm2 = np.sin(np2*2*np.pi*np.linspace(0,1,data_length,endpoint=False))
wfm3 = np.sin(np3*2*np.pi*np.linspace(0,1,data_length,endpoint=False))
wfm_trig = np.zeros(data_length)
wfm_trig[:8] = 1
plt.plot(wfm1)
plt.plot(wfm2)
plt.plot(wfm3)
plt.plot(wfm_trig)
```
